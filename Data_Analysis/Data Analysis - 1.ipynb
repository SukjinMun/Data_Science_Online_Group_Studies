## Data Frame

- data that is stored according to format - format that is saved in row x column square shaped "spreadsheet"
- according to how the data frame is usued differently, it is either called by spreadshet, table, excel and other names. 
- data saved in rows are stored as dictionaries
- data saved in columns are stored as lists
- rows are saved as lists in one data frame

```python
import pandas as pd

all_list = []
all_list.append({'date':'2022-03-19', 'dayofweek':'sat'})
all_list.append({'date':'2022-03-20', 'dayofweek':'sun'})
all_list.append({'date':'2022-03-21', 'dayofweek':'mon'})
all_list
```

```python
df = pd.DataFrame(all_list)
df
```

```python
# Creating data frame in different ways
name = ['jo', 'kim', 'lee', 'choi', 'park']
age = [22, 31, 55, 17, 27]

DataSet = list(zip(name, age))
df = pd.DataFrame(data=DataSet, columns=['name', 'age'])

#this prints out the list-ified zip
print(DataSet)

#this prints out the dataframe
df
```

```python
numbers = [1, 2, 3]
letters = ['a', 'b', 'c']

#The printed result is essentially zip(name, age) for the code right above
for pair in zip(numbers, letters):
    print(pair)
    
```

## Series
- 데이터가 순차적으로 나열된 1차원 배열 형태
- Index는 Value와 1:1 대응됨, Index는 Value의 위치를 나타내는 역할
- 키(k)와 값(v)이 '{k:v}' 형태로 짝을 이루는 파이썬의 Dictionary와 비슷한 구조를 갖음
- Dictionary의 k -> Series의 Index로,
- Dictionary의 v -> Series의 Value로 변환해서 사용

```python
dic = {'a':1, 'b':2, 'c':3}
sr = pd.Series(dic)
sr
```

```python
type(sr)
```

```python
ls = ['jo', 'kim', 'lee', 'choi', 'park']
sr = pd.Series(ls)
sr
```

## Data Frame
- 행과 열로 만들어지는 2차원 배열 구조
- k여러 개의 Series들이 모여서 Data Frame을 이루는 구조
- Data Frame의 열은 각각 Series 객체로, Data Frame은 여러 개의 Series들이 같은 행 Index를 기준으로 줄지어 결합된 2차원 행렬(matrix)임

```python
# Dict -> Data Frame 변환

dic = {'c0':[1, 2, 3], 
       'c1':[4, 5, 6], 
       'c2':[7, 8, 9], 
       'c3':[10, 11, 12], 
       'c4':[13, 14, 15]}

df = pd.DataFrame(dic)
df
```

```python
# List -> Data Frame 변환

df = pd.DataFrame([[20224943, '컴퓨터공학', 1], [20191234, '전자공학', 2]],
                 index=['jo', 'kim'], 
                 columns=['학번', '전공', '학년'])
df
```

```python
df.index = ['lee', 'park']
df.columns = ['number', 'major', 'grade']
```

```python
df
```

```python
df.rename(index={'lee':'jo'}, inplace=True)
df.rename(columns={'number':'stunumber'}, inplace=True)
df
```
